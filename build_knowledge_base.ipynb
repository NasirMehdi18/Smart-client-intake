```python
# build_vector_db.py → Sirf ek baar chalana hai
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# ←←← YE PATH CHANGE KAR LO (tera new PDF)
pdf_path = r"C:\Users\DELL\Desktop\lawAssistant\cleaned_no_toc.pdf"

print("Loading tera new PDF (76 pages)...")
loader = PyPDFLoader(pdf_path)
docs = loader.load()
print(f"Total pages loaded: {len(docs)}")  # 76 aayega

# Clean text
for doc in docs:
    doc.page_content = " ".join(doc.page_content.split())

print("Creating smart chunks...")
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(docs)
print(f"Total chunks: {len(chunks)}")

print("Loading embedding model...")
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cpu"}
)

print("Building & saving vector DB locally... (5-7 min)")
vectorstore = FAISS.from_documents(chunks, embeddings)

# Ye folder banega aur vector DB permanently save ho jayegi
os.makedirs("knowledge_base/vector_db", exist_ok=True)
vectorstore.save_local("knowledge_base/vector_db")

print("="*70)
print("HO GAYA BHAI! Vector DB permanently saved hai")
print("Ab har baar sirf app.py chalana – zero loading time!")
print("="*70)
```
